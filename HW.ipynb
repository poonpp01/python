6.จัดระดับความรุนแรงของแต่ละเขตในปี 2008

```python
import numpy as np
```

```python
df = pd.read_csv('Violent_Crimes_by_County__2006_to_2013.csv',index_col='County')
conditions = [
    (df['2008_Total_Violent_Crimes'] < 1000) & (df['2008_Total_Violent_Crimes'] > 300) ,
    (df['2008_Total_Violent_Crimes'] < 300) ,
    (df['2008_Total_Violent_Crimes'] >1000)]
choices = ['Medium', 'Low', 'High']
```

```python
df['Level'] = np.select(conditions, choices )
df[['2008_Total_Violent_Crimes', 'Level']]
```
